##### 20240525163216_BM_Soft_Transaction_Inquiry_01.pdf

```python
from img2table.document import PDF
from img2table.ocr import DocTR
import pandas as pd

file = PDF("sample/pdf/20240525163216_BM_Soft_Transaction_Inquiry_01.pdf")

ocr = DocTR(kw={"pretrained": True})

tables = file.extract_tables(ocr=ocr, borderless_tables=True, min_confidence=50).copy()
```

```python
result = pd.concat([tables[0].pop(2).df.fillna("")], ignore_index=True)
result
```
