## Bank Statement - PDF

##### 20240525162404_Rekening Koran 1.pdf


```python
from img2table.document import PDF
from img2table.ocr import DocTR
import pandas as pd

file = PDF("sample/pdf/20240525162404_Rekening Koran 1.pdf")

ocr = DocTR(kw={"pretrained": True})

tables = file.extract_tables(ocr=ocr, borderless_tables=True, min_confidence=50).copy()
```

```python
tables.items()
```

```python
result = pd.concat([tables[0].pop(0).df.fillna(""), tables[1].pop(0).df.fillna(""), tables[2].pop(0).df.fillna("")], ignore_index=True)
result
```
