## Bank Statement - PDF

##### 20240525095823_Contoh RK UOB.pdf

```python
from img2table.document import PDF
from img2table.ocr import DocTR
import pandas as pd

file = PDF("sample/pdf/20240525095823_Contoh RK UOB.pdf")

ocr = DocTR(kw={"pretrained": True})

tables = file.extract_tables(ocr=ocr, borderless_tables=True, min_confidence=50)
```

```python
tables.items()
```

```python
result = pd.concat([tables[0].pop(1).df.fillna(""), tables[1].pop(1).df.fillna(""), tables[2].pop(1).df.fillna("")], ignore_index=True)
result
```

```python
table_items = tables.items()

dataframes = []

for key, table_list in table_items:
    for table in table_list:
        dataframes.append(table.df.fillna(""))

result = pd.concat(dataframes, ignore_index=True)

result
```
