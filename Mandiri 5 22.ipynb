## Bank Statement - IMAGE

##### Mandiri 5 22.jpg

```python
from img2table.document import Image
from img2table.ocr import DocTR
import pandas as pd

file = Image("sample/image/Mandiri 5 22.jpg")

ocr = DocTR(kw={"pretrained": True})

tables = file.extract_tables(ocr=ocr, borderless_tables=True, min_confidence=80).copy()
```
